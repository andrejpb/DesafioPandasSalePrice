```python
import pandas as pd
```

## 1. Faça a leitura do arquivo casas_final.csv e do arquivo emails.csv

```python
df_casas = pd.read_csv('casas_final.csv')
df_casas.head()
```

```python
df_emails = pd.read_csv('emails.csv')
df_emails.head()
```

## 2. Cruze essas bases usando como chave a coluna "id_casa" para que tenhamos um único DataFrame para trabalhar.

```python
cruzamento = pd.merge(left=df_casas, right=df_emails, on='id_casa', how='inner')
cruzamento
```

## 3. A média de preço varia de acordo com a quantidade de banheiros? Mostre.

```python
a = cruzamento["FullBath"].value_counts() 
cruzamento.groupby(by=['FullBath'])['SalePrice'].mean()
```

## 4. Transforme a coluna "SalePrice" em uma coluna categórica, com as seguintes categorias: ['cheap', 'medium', 'expensive']. OBS: Crie uma nova coluna, não substitua a antiga.

```python
grupos_saleprice = ['cheap', 'medium', 'expensive']
df_categorico = cruzamento
df_categorico["SalePriceCategorico"] = pd.cut(df_categorico["SalePrice"], bins=3, labels=grupos_saleprice)
df_categorico
```
